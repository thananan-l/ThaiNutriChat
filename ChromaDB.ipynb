```python
# import
from langchain.document_loaders import TextLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma
```

```python
def create_doc_chunks(doc_path):

  # load the document and split it into chunks
  loader = TextLoader(doc_path)
  documents = loader.load()

  # split it into chunks
  text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
  docs = text_splitter.split_documents(documents)

  return docs
```

```python
docs = create_doc_chunks("/content/drive/MyDrive/My LLM Lab/DataSource/txt/อาหารสำหรับผู้ที่เป็นโรคไตเรื้อรัง.txt")

docs
```

```python
# create the open-source embedding function
embedding_function = HuggingFaceEmbeddings(model_name='intfloat/multilingual-e5-small')
```

```python
# persist_directory='./chroma_db'
persist_directory='/content/drive/MyDrive/My LLM Lab/chroma_db'
```

```python
# save to disk
saved_db = Chroma.from_documents(docs, embedding_function, persist_directory=persist_directory)
saved_db.persist()
```

```python
# load from disk
loaded_db = Chroma(persist_directory=persist_directory, embedding_function=embedding_function)
```

```python
# query it
query = "การแบ่งระยะของโรคไตเรื้อรัง"
docs = loaded_db.similarity_search_with_score(query)

docs[0]
```

```python
collection = loaded_db.get()

collection
```
